```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train.csv')
df_t=pd.read_csv('test.csv')
df
```

```python
df_t
```

```python
df=df.drop('Cabin',axis=1)
df=df.drop('Ticket',axis=1)
df=df.drop('Name',axis=1)
df_t=df_t.drop('Cabin',axis=1)
df_t=df_t.drop('Ticket',axis=1)
df_t=df_t.drop('Name',axis=1)
```

```python
df.isnull().sum()
```

```python
df_t.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mode()[0],inplace=True)
df_t['Age'].fillna(df_t['Age'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df_t=df_t.dropna()
```

```python
df['Sex']=df['Sex'].map({'male':0,'female':1})
df_t['Sex']=df_t['Sex'].map({'male':0,'female':1})
```

```python
df['Embarked']=df['Embarked'].map({'S':1,'C':2,'Q':3})
df_t['Embarked']=df_t['Embarked'].map({'S':1,'C':2,'Q':3})
```

```python
columns_names=['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
df
```

```python
df.info()
```

```python
df_t.info()
```

```python
x=df.drop('Survived',axis=1)
y=df['Survived']
x_test=df_t[columns_names]
```

```python
lr=LinearDiscriminantAnalysis()
lr.fit(x,y)
```

```python
y_pred=lr.predict(x_test)
lr.score(x_test,y_pred)
```

```python
x_test['Survived']=y_pred
x_test
```

```python
The_prediction=x_test
The_prediction['Sex']=The_prediction['Sex'].map({0:'male',1:'female'})
The_prediction['Embarked']=The_prediction['Embarked'].map({1:'S',2:'C',3:'Q'})
The_prediction
#The_prediction.to_csv('Titanic survived predicton')
```
